```python
from sklearn.model_selection import train_test_split #splitting for train and test
```

```python
X=df.drop('quality',axis=1)
```

```python
Y=df['quality']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg=LinearRegression()
```

```python
linreg.fit(X_train,Y_train)
```

```python
ypred=linreg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
rmse_test = (np.sqrt(mean_squared_error(Y_test, ypred)))
```

```python
rmse_test
```

```python
r2_test = r2_score(Y_test,ypred)
```

```python
r2_test
```

```python
#the rmse value of the basic linear regression model shows that it is an average model with an average r2 value.
```
